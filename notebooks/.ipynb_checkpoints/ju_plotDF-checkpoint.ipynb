# 1. Plot All Kinematics

```python
def plotDataFrames():
    for seletion in ["emu","mumu","mutau","mu4j","ee","emu2","etau","e4j"]:
        for nbjet in ["==1",">1"]:
            plotter = DFPlotter(seletion,nbjet)
            plotter.plotKinematics()
plotDataFrames()
```
