```python
import ROOT as root
import utility_common as common
from pylab import *
%matplotlib inline
```

```python
baseDir = common.getBaseDirectory()
f = root.TFile(baseDir+'data/root/arc.root')
eraLumin  = {"B":5.750,"C":2.573,"D":4.242,"E":4.025,"F":3.105,"G":7.576,"H":8.651}
cutFlow = {1:"all",2:"pass_Lumin_Mask",3:"pass_Trigger",4:"pass_muTrigger",5:"pass_eTrigger"}
```

```python
nTotal,means,stds = 0,[],[]
stage = 5

for era in ["B","C","D","E","F","G","H"]:
    n = f.Get("TotalEvents_electron_2016"+era).GetBinContent(stage)
    lumin = eraLumin[era]
    
    rate = n/lumin
    rateVar = rate**2 * (1/n + 0.025**2)
    means.append(rate)
    stds.append(rateVar**0.5)
    print("{:6.1f}+/-{:6.1f}".format(rate,rateVar**0.5))
    nTotal += n

aveAll = nTotal/35.9
```

```python
xpos = np.arange(7)
ypos = np.array([i/1e6 for i in means])
ystd = np.array([i/1e6 for i in stds ])
plt.figure(facecolor='w',figsize=(5,4))


ave = aveAll/1e6
aveStd = ave*0.025
plt.fill_between(np.arange(-1,8),ave+aveStd,ave-aveStd, alpha=0.5, color='C1',lw=0,label='Total Average')
plt.axhline(ave, color='C1',lw=2)

plt.errorbar(xpos,ypos,yerr=ypos*0.025,fmt='o',label='Era Average')
plt.ylabel(r'Event Rate [$10^6/fb^{-1}$]',fontsize=12)
plt.xticks(xpos,['16B','16C','16D','16E','16F','16G','16H'],fontsize=12)
plt.xlim(-0.5,6.5)

ylower, yupper = 0.8*ypos.min(),1.5*ypos.max()
yrange = yupper-ylower
plt.ylim(ylower,yupper)

plt.grid(linestyle='--',alpha=0.5)

plt.text(0, yupper-yrange*0.15, r'CMS $preliminary$',style='italic',fontsize='15',fontweight='bold')
plt.text(0, yupper-yrange*0.25, r'signle-$e$ '+cutFlow[stage], style='italic',fontsize='12')

plt.legend()
plt.savefig(baseDir+'plots/data_electron_{}.png'.format(cutFlow[stage]),dpi=300)
```
