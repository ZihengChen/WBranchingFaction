# 1. Make templates

```python
variation,tag = '', 'lltauVTight'
tp = DFTemplater(variation,'pickles_'+tag)
xShp,xShpVar,xCnt,xCntVar,yShp,yCnt = tp.makeTemplatesAndTargets()
```

```python
baseDir = common.getBaseDirectory()
common.makeDirectory(baseDir + "data/templates_{}/shaping/".format(tag))
common.makeDirectory(baseDir + "data/templates_{}/counting/".format(tag))

np.save(baseDir + "data/templates_{}/shaping/X_{}".format(tag,variation),xShp)
np.save(baseDir + "data/templates_{}/counting/X_{}".format(tag,variation),xCnt)

np.save(baseDir + "data/templates_{}/shaping/XVar_{}".format(tag,variation),xShpVar)
np.save(baseDir + "data/templates_{}/counting/XVar_{}".format(tag,variation),xCntVar)
if variation == '':
    np.save(baseDir + "data/templates_{}/shaping/Y_{}".format(tag,variation),yShp)
    np.save(baseDir + "data/templates_{}/counting/Y_{}".format(tag,variation),yCnt)
```

# 2. Plot Templates

```python
baseDir = common.getBaseDirectory()
X = np.load(baseDir + "data/templates_{}/shaping/X_.npy".format(tag))
Y = np.load(baseDir + "data/templates_{}/shaping/Y_.npy".format(tag))
```

```python
temp = np.concatenate(X,axis=1)
targ = np.concatenate(Y,axis=0)
# separate processes
mctt = np.sum(temp[0:5],axis=0)
mct  = np.sum(temp[5:10],axis=0)
# mcw  = np.sum(temp[10:15],axis=0)
mcz  = np.sum(temp[10:15],axis=0)
mcvv = np.sum(temp[15:20],axis=0)

mclist = [mcvv,mcz,mct,mctt]
```

```python
colorList = ["#a32020","#eb8c00","deepskyblue","mediumpurple"]
labelList = ["VV","Z","t","tt"]
chList = common.channelLsit()


fig, axes = plt.subplots(2, 1, sharex=True, facecolor='w',
                         gridspec_kw={'height_ratios':[3,1]},
                         figsize=(16,4))
fig.subplots_adjust(hspace=0)


c = np.arange(targ.size)
accum = np.zeros(targ.size)

# making plot

ax = axes[0]
for i in range(len(mclist)):
    height = mclist[i]
    ax.bar(c, height, bottom=accum, width=0.8, alpha=1, color=colorList[i],label=labelList[i])
    accum += height
ax.errorbar(c, targ,yerr=targ**0.5, c='k', fmt='.',markersize=10,lw=1, label='Data')
ax.grid(True,linestyle="--",alpha=0.5)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])
ax.set_yscale('log')
ax.set_ylim(10,5e5)
ax.text(0, 5e4, r'CMS $preliminary$',style="italic",fontsize="15",fontweight='bold')
ax.text(12, 5e4, tag, style="italic", fontsize=11)
ax.set_title("L=35.9/fb (13TeV)",loc="right",fontsize=14)

ax = axes[1]
ax.errorbar(c, targ/accum, yerr=targ**0.5/accum, c='k', fmt='.',markersize=10,lw=1)
ax.grid(True,linestyle="--",alpha=0.5)
ax.axhline(1,lw=1,color='k')
ax.set_ylim(0.5,1.5)
ax.set_xlim(c.min()-1., 1.1*c.max())
ax.set_xticks([])
ax.set_xlabel(r'$p^T_\tau$ [GeV], each channel range 20-80 GeV',fontsize=14)

plt.savefig('../../../plots/2020_tauID_prefit_{}.png'.format(tag),dpi=300)
```
