```python
from utility_dfcutter import *
from utility_dfplotter import *
import utility_common as common
%matplotlib inline
```

```python
variation,folderOfPickles = '','pickles_lltauVTight'
```

```python
selection,nbjet,njet = 'emutau','>=0','>=0'
mctt = DFCutter(selection,nbjet=nbjet,name='mctt',njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)
mct  = DFCutter(selection,nbjet=nbjet,name='mct',njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)
mctother  = DFCutter(selection,nbjet=nbjet,name='mctother',njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)
mcg  = DFCutter(selection,nbjet=nbjet,name='mcg',njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)

mcw  = DFCutter(selection,nbjet=nbjet,name='mcw',njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)
mcz  = DFCutter(selection,nbjet=nbjet,name='mcz',njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)
mcvv = DFCutter(selection,nbjet=nbjet,name='mcdiboson',njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)
data = DFCutter(selection,nbjet=nbjet,name="data", njet=njet,folderOfPickles=folderOfPickles).getDataFrame(variation)
```

```python
v,a,b,step,xl = 'tauGenFlavor',0,30,1, r'$\tau_h$ origin'
# v,a,b,step,xl = 'tauPt',0,100,5,r'$p^T_\tau$ [GeV]'
# v,a,b,step,xl = 'nJets',0,5,1, r'$n_j$'
# v,a,b,step,xl = 'dilepton_mass',0,250,5, r'$m_{e\mu}$'
dfList = [mcvv,mcg,mcz,mcw,mctother, mct,mctt,data]
sk = ASingleKinematicPlot(v,a,b,step,dfList)
sk.settingPlot(xl,
               ['VV',"$\gamma$","Z","W","t","tW","tt","data"], 
               ["C6","C5","C4","C3","C2","C1","C0",'k'],subtitle = r"$e\mu+\tau_h$ channel, $n_j\geq 0, n_b\geq 0$"
              )
            
sk.makePlot(normWithMC=True)
plt.savefig('../../../plots/misid/{}_{}_{}.png'.format(selection,v,folderOfPickles),dpi=300)
```
